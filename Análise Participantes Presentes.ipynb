## Bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

## Lendo CSV

```python
df_Presentes_2019 = pd.read_csv('Presentes2019.csv')
df_Presentes_2020 = pd.read_csv('Presentes2020.csv')
df_Presentes_2021 = pd.read_csv('Presentes2021.csv')
```

```python
#Colunas desnecessárias

df_Presentes_2019.drop(columns={'Unnamed: 0'},inplace=True)
df_Presentes_2020.drop(columns={'Unnamed: 0'},inplace=True)
df_Presentes_2021.drop(columns={'Unnamed: 0'},inplace=True)
```

## Análise Raça/Cor nos anos 2019, 2020 e 2021

```python
# Agrupamento de Participantes por Raça/Cor e Sexo

etnia_19 = df_Presentes_2019.groupby(['TP_COR_RACA','TP_SEXO']).agg(count=('TP_SEXO','count')).reset_index()
etnia_20 = df_Presentes_2020.groupby(['TP_COR_RACA','TP_SEXO']).agg(count=('TP_SEXO','count')).reset_index()
etnia_21 = df_Presentes_2021.groupby(['TP_COR_RACA','TP_SEXO']).agg(count=('TP_SEXO','count')).reset_index()
```

```python
plt.figure(figsize=(12,10))
sns.barplot(data=etnia_19, x='TP_COR_RACA', y='count', hue='TP_SEXO')
plt.xlabel('Cor/Raça')
plt.ylabel('Número de Participantes')
plt.legend(title='Sexo')
plt.title('Participantes 2019')
```

```python
plt.figure(figsize=(12,10))
sns.barplot(data=etnia_20, x='TP_COR_RACA', y='count', hue='TP_SEXO')
plt.xlabel('Cor/Raça')
plt.ylabel('Número de Participantes')
plt.legend(title='Sexo')
plt.title('Participantes 2020')
```

```python
plt.figure(figsize=(12,10))
sns.barplot(data=etnia_21, x='TP_COR_RACA', y='count', hue='TP_SEXO')
plt.xlabel('Cor/Raça')
plt.ylabel('Número de Participantes')
plt.legend(title='Sexo')
plt.title('Participantes 2021')
```

```python
# Agrupamento por raça/cor e média

nota_19 = df_Presentes_2019.groupby(['TP_COR_RACA','NU_ANO']).mean()['Média_Notas'].reset_index()
nota_20 = df_Presentes_2020.groupby(['TP_COR_RACA','NU_ANO']).mean()['Média_Notas'].reset_index()
nota_21 = df_Presentes_2021.groupby(['TP_COR_RACA','NU_ANO']).mean()['Média_Notas'].reset_index()
df_notas_p = pd.concat([nota_19,nota_20,nota_21]).reset_index()
```

```python
plt.figure(figsize=(15,12))
sns.lineplot(data=df_notas_p,x='NU_ANO',y='Média_Notas',hue='TP_COR_RACA')
plt.xlabel('Ano')
plt.ylabel('Média de Notas')
plt.legend(title='Etnia')
plt.title('Média das Notas por Etnia Entre os Anos')
```
