## Bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

## Importar CSV

```python
df = pd.read_csv("/content/drive/MyDrive/Faculdade/6° Período/Ciencia de Dados/IntroCD/02 - Analise exploratoria/DADOS2019/MICRODADOS_ENEM_2019.csv", sep=';', encoding='ISO-8859-1')
```

##Drop de colunas que não serão utilizadas


```python
df.drop(columns = ['NU_INSCRICAO','TP_ESTADO_CIVIL','TX_GABARITO_LC','TX_GABARITO_MT','TX_GABARITO_CN','TX_GABARITO_CH','TX_RESPOSTAS_CN',
                   'TX_RESPOSTAS_CH','TX_RESPOSTAS_LC','TX_RESPOSTAS_MT','NU_NOTA_COMP1','NU_NOTA_COMP2','NU_NOTA_COMP3','NU_NOTA_COMP4',
                   'NU_NOTA_COMP5','IN_TREINEIRO'], inplace=True)
```

```python
df.drop(columns = ['TP_ENSINO','TP_ANO_CONCLUIU','CO_MUNICIPIO_ESC','NO_MUNICIPIO_ESC','CO_UF_ESC','SG_UF_ESC','TP_DEPENDENCIA_ADM_ESC','TP_LOCALIZACAO_ESC', 'TP_SIT_FUNC_ESC',
                   'CO_MUNICIPIO_ESC','NO_MUNICIPIO_ESC','CO_UF_ESC','CO_UF_PROVA','CO_MUNICIPIO_PROVA','SG_UF_ESC','TP_DEPENDENCIA_ADM_ESC','TP_LOCALIZACAO_ESC', 'TP_SIT_FUNC_ESC',
                   'CO_PROVA_CN','CO_PROVA_CH','CO_PROVA_LC','CO_PROVA_MT','Q023','Q020','Q014','Q018','Q019','Q021','Q017','Q016','Q015','Q013','Q007'], inplace=True)
```

## Renomear colunas de perguntas

```python
df.rename(columns={'Q001': 'Estudo_Pai','Q002': 'Estudo_Mãe','Q003': 'Profissão_Pai','Q004': 'Profissão_Mãe','Q005': 'QT_Pessoas_Moradia',
                   'Q006': 'Renda_Familiar','Q008': 'QT_Banheiros','Q009': 'QT_Quartos','Q010': 'QT_Carros','Q011': 'QT_Motos',
                   'Q012': 'QT_Geladeira','Q022': 'QT_Celular','Q024': 'QT_Computador','Q025': 'Acesso_Internet'}, inplace = True)
```

## Separando dataset por alunos presentes, faltantes e anulados

```python
df_FaltasH = df[df.TP_PRESENCA_CH == 0]
df_FaltasN = df[df.TP_PRESENCA_CN == 0]
df_Faltas = pd.concat([df_FaltasH,df_FaltasN])

df_AnuladosH = df[df.TP_PRESENCA_CH == 2]
df_AnuladosN = df[df.TP_PRESENCA_CN == 2]
df_Anulados = pd.concat([df_AnuladosH,df_AnuladosN])

df_PresentesH = df[df.TP_PRESENCA_CH == 1]
df_PresentesN = df[df.TP_PRESENCA_CN == 1]
df_Presentes = pd.concat([df_PresentesH,df_PresentesN])
```

```python
df_Faltas = df_Faltas.drop_duplicates()
df_Anulados = df_Anulados.drop_duplicates()
df_Presentes = df_Presentes.drop_duplicates()
```

## Drop de linhas Nulas dos participantes presentes

```python
df_Presentes.dropna(inplace=True)
```

## Drop das colunas de presença

```python
df_Presentes.drop(columns = ['TP_PRESENCA_CN','TP_PRESENCA_CH','TP_PRESENCA_MT','TP_PRESENCA_LC'], inplace=True)
df_Faltas.drop(columns = ['TP_PRESENCA_CN','TP_PRESENCA_CH','TP_PRESENCA_MT','TP_PRESENCA_LC'], inplace=True)
df_Anulados.drop(columns = ['TP_PRESENCA_CN','TP_PRESENCA_CH','TP_PRESENCA_MT','TP_PRESENCA_LC'], inplace=True)
```

## Adicionando cor/raça, tipo de escola e faixa etária

---



```python
dic_raca = {0:'Não declarado',1:'Branca',3:'Parda',2:'Preta',4:'Amarela',5:'Indígena',6:'Sem informação'}
dic_tp_escola = {1:'Não respondeu',2:'Pública',3:'Privada',4:'Exterior'}
dic_Faixa_Etaria = {1:'Menor de 17 anos',2:'17 anos',3:'18 anos',4:'19 anos',5:'20 anos',6:'21 anos',7:'22 anos',8:'23 anos',9:'24 anos',10:'25 anos',
                    11:'Entre 26 e 30 anos',12:'Entre 31 e 35 anos',13:'Entre 36 e 40 anos',14:'Entre 41 e 45 anos',15:'Entre 46 e 50 anos',16:'Entre 51 e 55 anos',
                    17:'Entre 56 e 60 anos',18:'Entre 61 e 65 anos',19:'Entre 66 e 70 anos',20:'Maior de 70 anos'}
Dic_ST_EM = {1:'concluiu',2:'cursando',3:'cursando',4:'não concluiu'}
```

```python
df_Presentes['TP_COR_RACA'] = df_Presentes['TP_COR_RACA'].map(dic_raca)
df_Faltas['TP_COR_RACA'] = df_Faltas['TP_COR_RACA'].map(dic_raca)
df_Anulados['TP_COR_RACA'] = df_Anulados['TP_COR_RACA'].map(dic_raca)
```

```python
df_Presentes['TP_ESCOLA'] = df_Presentes['TP_ESCOLA'].map(dic_tp_escola)
df_Faltas['TP_ESCOLA'] = df_Faltas['TP_ESCOLA'].map(dic_tp_escola)
df_Anulados['TP_ESCOLA'] = df_Anulados['TP_ESCOLA'].map(dic_tp_escola)
```

```python
df_Presentes['TP_FAIXA_ETARIA'] = df_Presentes['TP_FAIXA_ETARIA'].map(dic_Faixa_Etaria)
df_Faltas['TP_FAIXA_ETARIA'] = df_Faltas['TP_FAIXA_ETARIA'].map(dic_Faixa_Etaria)
df_Anulados['TP_FAIXA_ETARIA'] = df_Anulados['TP_FAIXA_ETARIA'].map(dic_Faixa_Etaria)
```

```python
df_Presentes['TP_ST_CONCLUSAO'] = df_Presentes['TP_ST_CONCLUSAO'].map(Dic_ST_EM)
df_Faltas['TP_ST_CONCLUSAO'] = df_Faltas['TP_ST_CONCLUSAO'].map(Dic_ST_EM)
df_Anulados['TP_ST_CONCLUSAO'] = df_Anulados['TP_ST_CONCLUSAO'].map(Dic_ST_EM)
```

## Adicionando coluna com média de notas

```python
df_Presentes['Média_Notas'] = (df_Presentes['NU_NOTA_CN'] + df_Presentes['NU_NOTA_CH'] + df_Presentes['NU_NOTA_MT'] + df_Presentes['NU_NOTA_LC'] + df_Presentes['NU_NOTA_REDACAO']) / 5
```
